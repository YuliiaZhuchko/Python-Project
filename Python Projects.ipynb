##З датасету виделанo такі стовпці:
* alpha-3

Для зручності назви стовпців датасету було приведено до стандартизованого вигляду snake_case

```python

merged_events_countries = pd.merge(events_new, countrie, left_on='Country Code', right_on='alpha-3', how='left')

# З'єднуємо результат з таблицею products по стовпцях Product ID і id
merged_all = pd.merge(merged_events_countries, products, left_on='Product ID', right_on='id', how='left')

# Видалення непотрібних стовпців
merged_all = merged_all.drop(columns=[ 'alpha-3', 'id', 'sub-region'])


merged_all.rename(columns={
    'name': 'country_name',
    'Order ID': 'order_id',
    'Order Date': 'order_date',
    'Ship Date': 'ship_date',
    'Order Priority': 'order_priority',
    'Product ID': 'product_id',
    'Sales Channel': 'sales_channel',
    'Units Sold': 'units_sold',
    'Unit Price': 'unit_price',
    'Unit Cost': 'unit_cost',
    'region': 'region',
    'Country Code':'country_code',
    'item_type': 'product_type'
}, inplace=True)

# Перевірка результату
print(merged_all.head())
```

#Рохрахунок загальних показників:
* Загальна кількість замовлень:1330
* Загальна кількість країн:46
* Загальна кількість проданих одиниць товару:6586448.0
* Загальний дохід:1704628370.6500003
* Загальний прибуток:501842633.1200006
* Загальні витрати:1202785737.5299997
* Середній розмір замовлення:4952.216541353383

```python
total_orders = len(merged_all)

total_sales = sum(merged_all["unit_price"] * merged_all["units_sold"])
total_cost = sum(merged_all["unit_cost"] * merged_all["units_sold"])
total_profit = (total_sales - total_cost)

total_country_count = len(merged_all['country_name'].unique())

total_units_sold = sum(merged_all['units_sold'])

total_revenue = sum(merged_all['units_sold'] * merged_all['unit_price'])


avg_order_size = total_units_sold / total_orders

print(f"Загальна кількість замовлень:{total_orders}")
print(f"Загальна кількість країн:{total_country_count}")
print(f"Загальна кількість проданих одиниць товару:{total_units_sold}")
print(f"Загальний дохід:{total_revenue}")
print(f"Загальний прибуток:{total_profit}")
print(f"Загальні витрати:{total_cost}")
print(f"Середній розмір замовлення:{avg_order_size}")
```

#Аналіз продажів за днями тижня

```python
weekday_sales = merged_all.groupby('day_of_week')['revenue'].sum()
weekday_sales = weekday_sales.reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
weekday_sales.plot(kind='bar', figsize=(10,6))
plt.title('Продажі за днями тижня')
plt.ylabel('Продажі')
plt.xlabel('День тижня')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Перетворюємо колонку 'order_date' в datetime формат, якщо це не зроблено раніше
merged_all['order_date'] = pd.to_datetime(merged_all['order_date'])

# Додаємо колонку 'day_of_week'
merged_all['day_of_week'] = merged_all['order_date'].dt.day_name()

# Групуємо дані за типом товару і днем тижня, підсумовуємо продажі
sales_by_day_of_week = merged_all.groupby(['product_type', 'day_of_week'])['units_sold'].sum().unstack()

# Перерозподіляємо дні тижня за стандартним порядком
sales_by_day_of_week = sales_by_day_of_week.reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], axis=1)

# Візуалізація: Створюємо графік для кожного товару
sales_by_day_of_week.plot(kind='bar', figsize=(14,8), colormap='viridis', stacked=True)

# Додаємо підпис і заголовок
plt.title('Продажі товарів за днями тижня')
plt.ylabel('Продажі')
plt.xlabel('Тип товару')
plt.legend(title='День тижня', bbox_to_anchor=(1.05, 1), loc='upper left')

# Відображення графіка
plt.tight_layout()
plt.show()
print(sales_by_day_of_week)
```
